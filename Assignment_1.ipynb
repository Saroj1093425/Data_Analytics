```r
"Enter no of students in each branch::"
{
    cse = readline("Enter no of students in cse::");
    cst = readline("Enter no of students in cst::");
    ece = readline("Enter no of students in ece::");
    mech = readline("Enter no of students in mech::");
    bio = readline("Enter no of students in bio::");
}
cse = as.integer(cse);
cst = as.integer(cst);
ece = as.integer(ece);
mech = as.integer(mech);
bio = as.integer(bio);

sum = cse + cst + ece + mech + bio;
"sum of all stduents::"
sum
CSE = (cse/sum)*100;
CSE;
CST = (cst/sum)*100;
CST;
ECE = (ece/sum)*100;
ECE;
MECH = (mech/sum)*100;
MECH;
BIO = (bio/sum)*100;
BIO;
x <- c(CSE,CST,ECE,MECH,BIO)
mylabel <- c("CSE","CST","ECE","MECH","BIO")
pie(x,label = mylabel,main = "BRANCHES")
```
